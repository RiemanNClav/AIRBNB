Estadistica Bivariada con el objetivo de cunatificar la relación existente entre dos variables categóricas, establecer si existe independencia entre esta dos variables, la prueba Chi-cuadrado, nos ayudará a determinar estadisticamente si hay o no un grado de independencia entre dos variables categoricas. 

```python
import pandas as pd 
import os 
#libreria para el test chi cuadrado 
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
data = pd.read_csv('C:/Users/XMX5634/Documents/Introduccion Ciencia de Datos Aplicada/Estadistica Univariada/Ordenes.csv', sep = ';',
                encoding='latin-1')
```

```python
#construir una tabla de contingencia. 
data.head(5)
```

```python
# buscar los top 5 categorias de productos
categoria_top = data.groupby('nombre_categoria_producto').count().reset_index()[['nombre_categoria_producto', 'producto_id']]
categoria_top_lista = categoria_top.head()['nombre_categoria_producto']
categoria_top.head()
```

```python
# encontrar los departamentos top 5
departamento_top = data.groupby('departamento_vendedor').count().reset_index()[['departamento_vendedor', 'producto_id']]
departamento_top_lista = departamento_top.head()['departamento_vendedor']
departamento_top.head()
```

```python
#filtrar por los departamentos top.
data_top = data[data['nombre_categoria_producto'].isin(categoria_top_lista) & data['departamento_vendedor'].isin(departamento_top_lista)]
data_top.head(5)
```

```python
# realizar la tabla de contingencia.
tabla_contingencia = pd.crosstab(data_top['nombre_categoria_producto'],data['departamento_vendedor'], margins = True)
tabla_contingencia
```

Podemos ver que el departamento Antioquia en efecto es el que mayormente genera ventas de las categorias top, ahora vamos a extender el analisis por filas y columnas para identificar diferentes tipos de hipotesis. 

## Analisis Filas

 Hipotesis: es claro que la mayor cantidad de ventas se va a concentrar en Antioquia, por lo tanto la organización quiere analizar si la categorias Bebes, Carniceria y Celulares tienen alguna ciudad de preferencia de ventas

```python
# realizar la tabla de contingencia.
tabla_contingencia = pd.crosstab(data_top['nombre_categoria_producto'],data['departamento_vendedor'], margins = True, normalize = 'index')
tabla_contingencia
```

 Analisis: Para el caso de Bebes y Carniceria vemos que donde mas se vendio si fue un Antioquia.

## Analisis Columna

Hipotesis: Es necesario identificar en que departamento se tieien la mayor cantidad de ventas en las categoria de Carniceria. 

```python
pd.crosstab(data_top['nombre_categoria_producto'], data['departamento_vendedor'], margins = True, normalize = 'columns')
```

## Analisis de totales

tecnica estadistica para averiguar si las dos variables estan relacionadas y la manera de averiguarlos es mediante la distribución de porcentajes. Concretamente se analiza si la distribución de porcentajes de una variables se repiten por igual 
en las categorias de la otra variable. 


```python
pd.crosstab(data_top['nombre_categoria_producto'], data['departamento_vendedor'], margins = True, normalize = 'all')
```

# Prueba Chi Cuadrada 
la prueba chi, es una prueba estadistica que asume que las frecuencias observadas para una variable categorica coinciden con las frecuencias esperadas para la otra variable categorica. existencia o no de independencia entre dos variables

Hipotesis: Las ventas de las categorias de productos no estan relacionadas con el departamento que se venda. 

```python
# Establecer los valores del test chi cuadrado
chi, pval, dof, exp = chi2_contingency(tabla_contingencia)

print('p-valor es: ', pval)

#establecer la significancia del test
significancia = 0.05
p = 1 - significancia
#establecer el valor critico 
valor_critico = chi2.ppf(p, dof)
print('chi=%.6f, valor critico=%.6f\n' % (chi, valor_critico))

# segun el valor critico se toma la desición de rechazar o no la hipotesis
if chi > valor_critico:
    print(''' A un %.2f de nivel de significancia, se rechaza la hipotesis nula y se acepta H1.
    no son independientes, por lo tanto no estan correlacionados''' %(significancia))
else:
        print(''' A un %.2f de nivel de significancia, no se rechaza la hipotesis nula y se acepta H1.
    son independientes, no tienen relación''' %(significancia))
    
```

Podemos ver que no tienen algun tipo de relación las categorias y los sitios de venta.
