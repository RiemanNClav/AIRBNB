## Tipos de datos

```python
str(df.dtypes[0])
# Los datos categoricos son 'object'
```

```python
def variables(df):
    categorica=0
    categoricas = []
    numerica=0
    for col in df.columns:
        variable = df[[col]].dtypes[0]
        if str(variable) == 'object':
            categorica += 1
            categoricas.append(col)
        else:
            numerica += 1
    return ({'categoricas': categorica, 'numericas': numerica},categoricas)
        
        
```

```python
variables, categoricas = variables(df)
```
