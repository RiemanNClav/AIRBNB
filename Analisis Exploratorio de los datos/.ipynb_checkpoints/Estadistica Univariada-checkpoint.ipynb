## Tablas de Frecuencias

```python
df.isnull().sum()
```

```python
# frecuencia absoluta
frec_abs = df['neighbourhood'].value_counts().reset_index()
frec_abs.head()
```

```python
# frecuencia relativa 
frec_abs['porcentaje_neighbourhood'] = frec_abs['neighbourhood'].apply(lambda x: x/frec_abs['neighbourhood'].sum())
frec_abs
```

```python
# frecuencia absoluta y relativa acumulada con cumsum()
#absoluta acumulada 
frec_abs['acumulada_neighbourhood'] = frec_abs['neighbourhood'].cumsum()
# relativa acumulada
frec_abs['acumulada_porcentaje_neighbourhood'] = frec_abs['porcentaje_neighbourhood'].cumsum()
frec_abs
```

```python
# frecuencia bivariada 
pd.crosstab(index = df['room_type'], columns = df['neighbourhood'])
```

## Medidas de DispersiÃ³n 

```python
# Desviacion Estandar 
df['price'].std()
```

```python
# Rango Intercuartilico 
df['price'].quantile(0.75) - df['price'].quantile(0.25)
```

```python
# Rango 
df['price'].max() - df['price'].min()
```

```python
# Coeficiente de VariaciÃ³n 

df['price'].std()/ df['price'].mean()
```
